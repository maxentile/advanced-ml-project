```python
from scipy.cluster import hierarchy
from scipy.spatial import distance
from scipy.spatial.distance import squareform,pdist
```

```python
from sklearn import datasets
data = datasets.load_digits()
X = data.data
Y = data.target
```

```python
import numpy as np
import numpy.random as npr
n = 1000
npr.seed(0)
rand = npr.rand(len(X))
ind = sorted(np.arange(len(X)),key=lambda i:rand[i])
X_s = X[ind][:n]
Y_s = Y[ind][:n]
```

```python
c = hierarchy.linkage(X_s,'single')
```

```python
from time import time
```

```python
X_ = []
t = time()
for i in range(10):
    tpe = TPE()
    X_.append(tpe.fit_transform(X_s))
    print(i)
    print(time() - t)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(X,Y,title='Embedding'):
    plt.scatter(X[:,0],X[:,1],c=Y,linewidths=0)
    plt.title(title)
```

```python
for x in X_:
    plot(x,Y_s)
    plt.figure()
```

```python
def compare_pairwise_distances(X1,X2):
    return np.sum((pdist(X1) - pdist(X2))**2)
```

```python
D = [hierarchy.cophenet(c) for c in C]
```
